```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# This ensures your plots appear inside the notebook
%matplotlib inline
```

```python
# Create a range of 30 days
dates = pd.date_range(start='2025-01-01', periods=30)

# Create random sales data using NumPy
# np.random.randint(min, max, size)
sales_values = np.random.randint(200, 1000, size=30)

# Put it all into a Pandas Table
df = pd.DataFrame({'Date': dates, 'Sales': sales_values})

# Add a 'Trend' column using a Pandas rolling average
df['Trend'] = df['Sales'].rolling(window=5).mean()

print("--- First 5 rows of our Data ---")
print(df.head())
```

```python
# Set the size of the graph
plt.figure(figsize=(10, 5))

# Plot the actual daily sales
plt.plot(df['Date'], df['Sales'], label='Daily Sales', color='skyblue', marker='o')

# Plot the trend line (Calculated with Pandas)
plt.plot(df['Date'], df['Trend'], label='5-Day Trend', color='red', linewidth=2)

# Use NumPy to find the average and draw a "Goal" line
avg_sales = np.mean(df['Sales'])
plt.axhline(avg_sales, color='green', linestyle='--', label=f'Avg: {avg_sales:.0f}')

# Add Labels and Title
plt.title('Sales Performance Analyzer', fontsize=16)
plt.xlabel('Date')
plt.ylabel('Sales ($)')
plt.legend()
plt.grid(True, alpha=0.3)

# Show the plot
plt.show()
```
