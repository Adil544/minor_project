```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Creating a synthetic dataset of 100 students
np.random.seed(42)
data = {
    'Student_ID': range(1, 101),
    'Name': [f'Student_{i}' for i in range(1, 101)],
    'Math': np.random.randint(40, 100, 100),
    'Science': np.random.randint(35, 100, 100),
    'English': np.random.randint(50, 98, 100)
}

df_generated = pd.DataFrame(data)
df_generated.to_csv('student_marks.csv', index=False)
print("File 'student_marks.csv' created successfully!")
```

```python
# Read the marks from CSV
df = pd.read_csv('student_marks.csv')

# Display the first 5 rows
df.head()
```

```python
# Calculate statistics for each subject
stats = pd.DataFrame({
    'Average': df[['Math', 'Science', 'English']].mean(),
    'Maximum': df[['Math', 'Science', 'English']].max(),
    'Minimum': df[['Math', 'Science', 'English']].min()
})

print("--- Subject-wise Performance ---")
print(stats)

# Calculating a total score for comparison
df['Total'] = df['Math'] + df['Science'] + df['English']
```

```python
# Sort students by total marks and take the top 10
top_10 = df.sort_values(by='Total', ascending=False).head(10)

# Plotting
plt.figure(figsize=(12, 6))
plt.bar(top_10['Name'], top_10['Total'], color='skyblue', edgecolor='navy')

plt.title('Top 10 Students by Total Marks', fontsize=15)
plt.xlabel('Student Name', fontsize=12)
plt.ylabel('Total Marks', fontsize=12)
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```
